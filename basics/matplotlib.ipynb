```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Basic Graph

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.title("first graph")
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y)
plt.title("first graph",fontdict={'fontname':'Comic Sans MS','fontsize':20})
plt.xlabel('x axis')
plt.ylabel('y axis')
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y)
plt.title("first graph")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y,label='2x',color='red',linewidth=8,marker='.',linestyle='--',markersize=10,markeredgecolor='blue',markerfacecolor='yellow')
plt.title("first graph")
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.legend()
```
