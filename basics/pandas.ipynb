```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df=pd.read_csv(url)
print(df.tail(2))
```

```python
##read headers
print(df.columns)
##read each column
print(df[['Name','Survived','Cabin']])
##read each row
print(df.iloc[1:4])
##read a specific location
print(df.iloc[2,2])

```

```python
print(df.iloc[2,2])
```

```python
df.sort_values(['Name'],ascending=False)
```

```python
df.head(5)
```

```python
df['Total']=df['SibSp']+df['Parch']
df.head()
```

```python
df=df.drop(columns=['Total'])
df.head()
```

```python
df['Total']=df.iloc[:,5:8].sum(axis=1)
df.head()
```

```python
cols=list(df.columns.values)
df=df[cols[0:5]+[cols[-1]]+cols[5:11]]
df.head()
```

saving our csv

```python
df.to_csv('modified.csv',index=False)
```

```python
df.loc[df['Name'] == 'Braund, Mr. Owen Harris', 'Name']='Varshika'
df.head()
```

```python
df.loc[df['Total']>20,'Total']=30
df.head()
```
